1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
df = pd.read_csv('previsao_de_renda.csv')
````

````python
df
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda", hue = 'tipo_renda',
                   data=df, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda", hue = 'tipo_renda',
                   data=df, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

A renda média de pessoa que possuem imóvel e veículo é estatisticamente maior do que aquelas que não possuem
